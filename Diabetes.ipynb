1. TENTUKAN LIBRARY 


```python
#Masukkkan semua librari yang diburuhkan dalam program
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

2. LOAT DATA SET

```python
#untuk memanggil data csv
datadiabetes = pd.read_csv ('diabetes.csv')
```

```python
#perintah untuk menampilkan data pada file diabetes.csv
print ('\nData Penderita Diabetes')
datadiabetes
```

```python
#perintah menampilkan Jumlah Baris dan Kolom pada Data
datadiabetes.shape
```

```python
#perintah menampilkan Jumlah dari Variabel Label
datadiabetes['Hasil'].value_counts()
```

```python

#perintah memisahkan Data dengan Label menjadi Variabel X dan Y
X= datadiabetes.drop (columns='Hasil',axis=1)
Y= datadiabetes['Hasil']
```

```python
#Perintah menampilkan data pada variabel X
print (X)
```

```python
#Perintah menampilkan data pada variabel X
print (Y)
```

3. STANDARISASI DATA

```python
#Karena nilainya berbeda-beda kita membutuhkan skaler untuk melakukan transpormasi data
scaler= StandardScaler()
scaler.fit(X)
standarisasi_data = scaler.transform(X)
print(standarisasi_data)
```

```python
#perintah melakukan definisi terhadap X dan Y
X= standarisasi_data
Y = datadiabetes['Hasil']

#perintah menampilkan variabel X dan Y
print(X)
print(Y)
```

4. MEMISAHKAN DATA TRAINING DAN DATA TESTING

```python
#perintah untuk memisahkan data menjadi data testing 20% dan data training 80% dengan Y sebagai variabel 
X_training, X_testing, Y_training, Y_testing= train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
#perintah untuk menampilkan jumlah data testing dan training yang sudah dipisahkan
print(X.shape,X_training.shape, X_testing.shape)
```

5. PERMODELAN DENGAN SVM

```python
#identifikasi untuk metode svm
kelas= svm.SVC(kernel='linear')
```

```python
kelas.fit(X_training,Y_training)
```

6. MEMBUAT MODEL EVALUSI UNTUK MENGUKUR TINGKAT AKURASI

```python
#perintak untuk mendeteksi prediksi dengan data training
prediksi_X_training =kelas.predict(X_training)
training_data = accuracy_score(prediksi_X_training,Y_training)
```

```python
#perintah menampilkan besarnya nilai akurasi data traning
print('Akurasi Data Traning Adalah:',training_data)
```

```python
#perintak untuk mendeteksi prediksi dengan data testing
prediksi_X_testing =kelas.predict(X_testing)
testing_data = accuracy_score(prediksi_X_testing,Y_testing)
```

```python
print('Akurasi Data Testing Adalah:',testing_data)
```

7. MEMBUAT MODEL PREDIKSI

```python
input_data_uji = (1, 85, 66, 29, 0, 26.6, 0.351, 31)
input_data_as_numpy_array =np.array(input_data_uji)
input_data_reshape= input_data_as_numpy_array.reshape(1,-1)
std_data=scaler.transform(input_data_reshape)
print(std_data)

prediksi =kelas.predict(std_data)
print(prediksi)
if(prediksi[0]== 0):
    print ('Tidak Menderita Diabetes')
else:
    print('Menderita Diabetes')
```
